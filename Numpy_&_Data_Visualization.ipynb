```python
days=['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
steps=[6745,8765,7456,4567,8799,3456,11567]
lwsteps=[8745,6784,11457,5964,7845,7945,13457]
plt.plot(days,steps,'o-g')
```

```python
plt.plot(days,lwsteps,'o-m')
```

```python
plt.title("Steps Count - Last Week & This Week")
```

```python
days=['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
steps=[6745,8765,7456,4567,8799,3456,11567]
lwsteps=[8745,6784,11457,5964,7845,7945,13457]
plt.plot(days,steps,'o-g')
plt.plot(days,lwsteps,'o-m')
plt.title("Steps Count - Last Week & This Week")
```

```python
days=['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
steps=[6745,8765,7456,4567,8799,3456,11567]
lwsteps=[8745,6784,11457,5964,7845,7945,13457]
plt.plot(days,steps,'o-g')
plt.plot(days,lwsteps,'o-m')
plt.title("Steps Count - Last Week & This Week")
plt.xlabel('Days Of the week')
plt.ylabel('Steps Walked')
```

```python
days=['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
steps=[6745,8765,7456,4567,8799,3456,11567]
lwsteps=[8745,6784,11457,5964,7845,7945,13457]
plt.plot(days,steps,'o-g')
plt.plot(days,lwsteps,'o-m')
plt.title("Steps Count - Last Week & This Week")
plt.xlabel('Days Of the week')
plt.ylabel('Steps Walked')
plt.grid(True)
```

```python
days=['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
steps=[6745,8765,7456,4567,8799,3456,11567]
lwsteps=[8745,6784,11457,5964,7845,7945,13457]
plt.plot(days,steps,'o-g')
plt.plot(days,lwsteps,'o-m')
plt.title("Steps Count - Last Week & This Week")
plt.xlabel('Days Of the week')
plt.ylabel('Steps Walked')
plt.grid(True)
plt.legend(['Current Week','Last Week'])
```

```python
plt.figure(figsize=(10,5))
age=[34,36,41,43,28]
name=['VK','SKY','MSD','ST','HP']
plt.bar(name,age)
```

```python
plt.figure(figsize=(10,5))
name=['Shivam','Suryansh','Shrikant','Vivek']
ot=[29,48,32,46]  #y1
ct=[22,34,15,28]  #y2
plt.bar(name,ot,label='Open-Tickets',width=0.5,color='k')
plt.bar(name,ct,label='Close-Tickets',width=0.5,color='b')
plt.xlabel('Name')
plt.ylabel('Tickets')
plt.legend()
```

```python
age=[34,36,41,43,28]
name=['VK','SKY','MSD','ST','HP']
colors=['g','r','b','k','m']
plt.pie(age,labels=name,colors=colors,startangle=45)
```
